### **일괄 추가**

- 생각해보니까 어떻게 문장 속에 있는 UNK 토큰에 해당하는 원본 단어를 찾지...?
- UNK 단어 하나를 붙잡고 알아보자
    - encode -> decode 한 결과와 원래 단어/문장을 비교
    - 원래 단어/문장에 있던 단어가 decode한 결과에 없으면 UNK 토큰!

```python
e1_UNK = []
e2_UNK = []
sentence_UNK = []
```

```python
UNK = []
```

```python
e1 = list(dataset[2])
e2 = list(dataset[5])
sentence = list(dataset[1])
```

```python
e1_UNK = []
e2_UNK = []
sentence_UNK = []

for i in range(len(e1)):
    e1_tok = tokenizer.decode(tokenizer.encode(e1[i], add_special_tokens=False)).split()
    e2_tok = tokenizer.decode(tokenizer.encode(e2[i], add_special_tokens=False)).split()
    sentence_tok = tokenizer.decode(tokenizer.encode(sentence[i], add_special_tokens=False)).split()
    
    if "[UNK]" in e1_tok:
        for e in e1[i].split():
            if e not in e1_tok:
                e1_UNK.append(e)
    if "[UNK]" in e2_tok:
        for e in e2[i].split():
            if e not in e2_tok:
                e2_UNK.append(e)
                print(i)
    if "[UNK]" in sentence_tok:
        for s in sentence[i].split():
            if s not in sentence_tok:
                sentence_UNK.append(s)
```

```python
# 원하는 entity 네임을 지정할 수도 있게 하고 싶은데 그러면 파라미터가 너무 많아진다
def add_entity_tokens(sentence, a1, a2, b1, b2):
    new_sentence = None
    if a1 > b1: # b1 먼저
        new_sentence = sentence[:b1] + "[E02]" + sentence[b1:b2+1] + "[/E02]" + sentence[b2+1:a1] + "[E01]" + sentence[a1:a2+1] + "[/E01]" + sentence[a2+1:]
    else: # a1 먼저
        new_sentence = sentence[:a1] + "[E01]" + sentence[a1:a2+1] + "[/E01]" + sentence[a2+1:b1] + "[E02]" + sentence[b1:b2+1] + "[/E02]" + sentence[b2+1:]
    return new_sentence
```
